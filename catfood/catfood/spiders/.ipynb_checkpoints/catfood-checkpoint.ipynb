```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def p2f(x):
    return float(x.strip('%'))
```

```python
#df = pd.read_csv('food_2019-06-27T22-32-47.csv', index_col='name').dropna(axis=1)
df = pd.read_csv('food_2019-06-27T22-32-47.csv').dropna(axis=1)
```

```python
df
```

```python
df.info()
```

```python
p_cols = [col for col in df.columns if '%' in df[col][0]]
p_cols
```

```python
paw_cols = ['ingredients_paws', 'nutrition_paws']
```

```python
for col in p_cols:
    df[col] = df[col].str.rstrip('%').astype('float')
```

```python
for col in paw_cols:
    df[col] = df[col].str[1].astype('float')
```

```python
df['dma_cals'] = df['dma_cals'].map(lambda x:x.split('/')[0].replace(',', '')).astype('float')
```

```python
df.info()
```

```python
df.head()
```

```python
g_brand = df.groupby('brand').mean().sort_values('nutrition_paws', ascending=False)
g_brand
```

```python
sns.heatmap(df[df['dry_wet'] == 'dry'].corr(), square=True)
```

```python
n_paws = X_n_paws = df[[col for col in df.columns if "dma" in col]]
n_paws = pd.concat([n_paws, df['nutrition_paws']], axis=1)
n_paws.head()
```

```python
sns.heatmap(n_paws.corr(), cmap='seismic', square=True, annot=True)
```

```python
n_paws = n_paws.drop(['dma_fiber', 'dma_fat', 'dma_ash'], axis=1)
```

```python
n_paws['nutrition_paws'].value_counts()
```

```python
X_n_paws = n_paws.drop('nutrition_paws', axis=1)
X_n_paws.head()
```

```python
y_n_paws = n_paws['nutrition_paws']
y_n_paws.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(multi_class='multinomial',
                       solver='newton-cg')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_n_paws, y_n_paws, test_size=0.33, random_state=42, stratify=y_n_paws)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
sns.heatmap(data=confusion_matrix(y_test, y_pred),
           square=True,
           annot=True,
           fmt='d')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(gamma='auto')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
sns.heatmap(data=confusion_matrix(y_test, y_pred),
           square=True,
           annot=True,
           fmt='d')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[10, 1000, 10000, 100000], 'gamma':[10e-1, 10e-2, 10e-3, 10e-4]}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, verbose=10, cv=5)
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
y_pred = grid.predict(X_test)
```

```python
sns.heatmap(data=confusion_matrix(y_test, y_pred),
           square=True,
           annot=True,
           fmt='d')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
